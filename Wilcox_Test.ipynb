###A researcher is studying the effectiveness of a new meditation-based stress reduction program on people diagnosed with high anxiety levels. After completing the program, participants were monitored, and their number of calm days (days without reported anxiety symptoms) in the following 6 months was recorded.

The number of calm days for 13 participants is given below:

124, 42, 25, 45, 412, 51, 1112, 46, 103, 876, 146, 340, 396

```r
claim_days<-c(124, 42, 25, 45, 412, 51, 1112, 46, 103, 876, 146, 340, 396)
```

```r
claim_days
```

### Drawing the Boxplot and Q-Q Plot to Identify Outliers and Assess Normality

```r
boxplot(
  claim_days,
  main="Stress Reduction Programme",
  ylab="claim_days",
  col="cyan"

)
```

```r
qqnorm(claim_days,main="Q-Q plot")
qqline(claim_days,col="red")
```

```r
shapiro.test(claim_days)
```

### Hypothesis Statement:
Check whether the median number of calm days exceeds 100

## H0 : median <= 100
##    H1 : median > 100

## Perform the Wilcox Rank test for above Data set

```r
wilcox.test(claim_days,mu=100,alternative="greater")
```

## WE DO NOT REJECT NULL HYPOTHESIS.
## Conclusion ---> We do not have enough evidence to say that claim days exceeds 100 days.
